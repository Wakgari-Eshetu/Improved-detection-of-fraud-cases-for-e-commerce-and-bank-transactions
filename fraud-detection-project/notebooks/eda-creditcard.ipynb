```python
# ===============================
# EDA: Credit Card Fraud Dataset
# ===============================

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# -------------------------------
# 1. Load Dataset
# -------------------------------
file_path = "../data/raw/creditcard.csv"
df = pd.read_csv(file_path)
print("Dataset shape:", df.shape)
df.head()
```

```python
# -------------------------------
# 2. Basic Info & Missing Values
# -------------------------------
print(df.info())
print("\nMissing values:\n", df.isnull().sum())
```

```python
# -------------------------------
# 3. Target Class Distribution
# -------------------------------
target_col = 'Class'
class_counts = df[target_col].value_counts()
print(class_counts)

plt.figure(figsize=(6,4))
sns.countplot(x=target_col, data=df)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.show()
```

```python
# -------------------------------
# 4. Summary Statistics
# -------------------------------
df.describe().T
```

```python
# -------------------------------
# 5. Correlation Heatmap
# -------------------------------
plt.figure(figsize=(12,10))
corr = df.corr()
sns.heatmap(corr, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# -------------------------------
# 6. Feature Distribution
# -------------------------------
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
num_cols.remove(target_col)

plt.figure(figsize=(15,20))
for i, col in enumerate(num_cols):
    plt.subplot(len(num_cols)//3+1, 3, i+1)
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# 7. Boxplots to Check Outliers
# -------------------------------
plt.figure(figsize=(15,20))
for i, col in enumerate(num_cols):
    plt.subplot(len(num_cols)//3+1, 3, i+1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# 8. Fraud vs Non-Fraud Comparison
# -------------------------------
plt.figure(figsize=(15,20))
for i, col in enumerate(num_cols):
    plt.subplot(len(num_cols)//3+1, 3, i+1)
    sns.kdeplot(df.loc[df[target_col]==0, col], label="Non-Fraud", fill=True)
    sns.kdeplot(df.loc[df[target_col]==1, col], label="Fraud", fill=True)
    plt.title(col)
plt.legend()
plt.tight_layout()
plt.show()
```
